```python
#input file names from user
file1=input("Enter the source file:")
file2=input("Enter the destination file:")
#open source and destination files
f1=open(file1,"r")
f2=open(file2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0: #skip even lines
        continue
    f2.write(line)
#close both filesfiles
f1.close()
f2.close()
print("Odd lines copied succesfully from",file1,"to",file2)
```

```python
import csv
#input csv file name
fname=input("Enter csv file name:")
#open csv file in read mode
f1=open(fname,"r")
#create csv reader object
reader=csv.reader(f1)
#read and print each row
for row in reader:
    print(row)
#close the file
f1.close()
```

```python
import csv 
#input csv file name
fname=input("Enter the csv file name:")
#open the csv file
with open(fname,"r")as f1:
    reader=csv.DictReader(f1)
    print("\nDepartment Details:")
    print("........")
    for row in reader:
        print("DeptId:",row['DeptId'],"\t DeptName:",row['DeptName'])
```
