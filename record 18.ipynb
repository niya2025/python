```python
#time overloading () 
class time:#at
    #campare areas 
    def __init__(self,hour,minute,second):#constructer__init__
        self.hour=hour
        self.minute=minute
        self.second=second
    def __add__(self,other):
        s=self.second+other.second
        m=self.minute+other.minute
        h=self.hour+other.hour
        if s>=60:
            s+s%60
            m=m+1
        if m>=60:
            m=m%60
            h=h+1
        print("added time:")
        print("hour:",h,"minute:",m,"second:",s)
hour1=int(input("ENTER THE HOUR:"))
minute1=int(input("ENTER THE MINUTE:"))
second1=int(input("ENTER THE SECOND:"))

hour2=int(input("ENTER THE HOUR:"))
minute2=int(input("ENTER THE MINUTE:"))
second2=int(input("ENTER THE SECOND:"))
#obj
t1=time(hour1,minute1,second1)
t2=time(hour2,minute2,second2)
t3=t1+t2
```

```python
#rectangle
class Rectangle:
    def __init__(self, l, w):
        self.length = l
        self.width = w
        self.area = l * w

    def __lt__(self, other): 
        if self.area < other.area:
            print("Rectangle 2 is greater")
        else:
            print("Rectangle 1 is greater")
l1 = int(input("Enter length of rectangle 1: "))
a1 = int(input("Enter width of rectangle 1: "))
l2 = int(input("Enter length of rectangle 2: "))
a2 = int(input("Enter width of rectangle 2: "))
r1 = Rectangle(l1, a1)
r2 = Rectangle(l2, a2)
r1 < r2
```
